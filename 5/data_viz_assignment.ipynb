```python
drones = pd.read_csv("data/drones.csv")
drones.info()
```

```python
# Histogram - A frequency distribution shows how often each different value in a set of data occurs.
# Using a histogram we see the distribution of deaths and injuries caused drone attacks

plt.hist([drones["Maximum Total Killed"], drones["Max injured"]], label = ['Deaths', 'Injuries'])
plt.grid(True) 
plt.legend()
plt.xlabel("Number of Attacks")
plt.ylabel("Number of Casualities")
plt.title('Histogram of casualities')
```

```python
# Scatter plots are used to plot data points on a horizontal and a
# vertical axis in the attempt to show how much one variable is affected by another. 

x = drones["Strike ID"]
y = drones["Maximum Total Killed"]
z = drones["Max injured"]

fig = plt.figure(figsize = (15, 10))
ax = fig.add_subplot(1, 1, 1)
ax.grid(True)
ax.set_title("Scatter Plots for Strike Id vs Max Killed & Strike Id vs Max Injured", fontsize='15')
ax.set_xlabel("Strike Id", fontsize='15')
ax.set_ylabel("Casualities", fontsize='15')
ax.scatter(x, y, c = 'g')
ax.scatter(x, z, c = 'b', marker = 'x')
```

```python
# Scatter Matrix:  is a statistic that is used to make estimates of the covariance matrix
# Draw a matrix of scatter plots.
pd.scatter_matrix(drones[['Maximum Total Killed', 'Maximum civilians killed']], figsize = (10, 10))
```

Trellis plotting
-----------------
+ arranges data in a rectangular grid by values of certain attributes (Pandas help)
+ splits plotted data series up into cells of a grid. Putting series into separate mini-charts eliminates the clutter that occurs when too many points are plotted together in the same chart, while still displaying comparisons and trends.
+ \note - Dot plots are a useful form trellis plotting, but are available only through gplot 

Example: Dot plots In the drones data set. For a given "Area" and "Strike Id" Plot the various number of casualities.

```python
titanic = pd.read_csv("data/titanic.csv")
titanic.info()
```

```python
import pandas.tools.rplot as rplot
```

```python

# http://pandas.pydata.org/pandas-docs/stable/rplot.html#scales
plt.figure(figsize = (10, 10))

plot = rplot.RPlot(titanic, x ='fare', y ='age')
plot.add(rplot.TrellisGrid(['sex', 'survived']))
plot.add(rplot.GeomHistogram())
plot.render(plt.gcf())

# UNABLE TO UNDERSTAND WHY THE PLOTS ARE REPEATED TWICE.
```
